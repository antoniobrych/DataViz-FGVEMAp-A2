```python
#Convertendo as datas de vencimento dos graos, para um padrao mais simples.
import dateparser

df_coffee['Expiration'] = df_coffee['Expiration'].apply(lambda x: dateparser.parse(x).strftime('%m/%y') if isinstance(x, str) else x)
```

```python

"""
esse codigo faz algo interessante, 

as altitudes estao como dados categoricos e por tanto operacoes estatisticas
nao podem ser efetuadas com esses dados, o q eu fiz foi pegar os numeros 
converter pra floar e se eles estao entre strings como 1290 - 2033, tirar a media
deles. 


"""
import re

def convert_altitude(value):
    if isinstance(value, str) and '-' in value:
        range_values = re.findall(r'\d+', value)
        range_values = [float(v) for v in range_values]
        return np.mean(range_values)
    else:
        try:
            return float(value)
        except ValueError:
            return np.nan

df_coffee['Altitude'] = df_coffee['Altitude'].apply(convert_altitude).astype(float)
```

```python
#verificando os paises que contam com altitudes com valor NaN
nan_altitude_countries = np.unique(df_coffee[df_coffee['Altitude'].isna()==True]['Country of Origin'].values)
nan_altitude_countries
```

```python
#calculando a media das altitudes que nao sao Nan
fill_nan_value = df_coffee[df_coffee['Country of Origin']==nan_altitude_countries[0]]['Altitude'].dropna().mean()
```

```python
#Valor
fill_nan_value
```

```python
#preenchendo o valor NaN, mas ainda nao substituindo no df
c1 = df_coffee[df_coffee['Country of Origin']==nan_altitude_countries[0]]['Altitude'].fillna(fill_nan_value)
```

```python
#calculando a media das altitudes que nao sao Nan
#agora do outro pais (Guatemala)
fill_nan_value = df_coffee[df_coffee['Country of Origin']==nan_altitude_countries[1]]['Altitude'].dropna().mean()
fill_nan_value
```
